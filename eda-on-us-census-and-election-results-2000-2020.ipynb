# Feature Relationships

(See how features relate with each other)
* Scatterplot
* Heatmap Correlation
* Pairplot
* Groupby comparisons

```python
ax = sns.scatterplot(data=df_2020, 
                x='Avg_Income',
                y='Avg_Mortgage', 
                hue='Masters_or_Pro_Pct')
ax.set_title('Average Income vs. Average Mortgage and Relation with Master Degree')
plt.show()
```

```python
#Use pair plot to see how each variable interact with each other
sns.pairplot(data=df, vars=['Avg_Income', 'Highschool_or_Below_Pct', 
                            'Masters_or_Pro_Pct', 'Doc_Degree_Pct',
                            'Employed_Pct', 'Not_in_Labor_Pct', 'Unemployeed_Pct'],
                        hue='Year')
plt.show()
```

```python
#See the correlation between variables from corr table
df_corr = df[['Avg_Income', 'Highschool_or_Below_Pct', 'Masters_or_Pro_Pct',
   'Doc_Degree_Pct','Employed_Pct', 'Not_in_Labor_Pct', 
   'Unemployeed_Pct']].dropna().corr()
df_corr
```

```python
#Visualize the correlation between variables
sns.heatmap(df_corr, annot=True)
```

# Ask Questions about the Data

What are the counties with the highest Masters or Professional Certificate?

```python
ax = df.groupby('County_Fips')['Masters_or_Pro_Pct'] \
    .agg(['mean','count']) \
    .sort_values(by=['mean'],ascending=False)['mean'] \
    .head(3) \
    .plot(kind='barh', figsize=(12,5), title='Average Master or Professional Percent by County')
ax.set_xlabel('Average Master or Professional Percent')
plt.show()

##County 51013, 51510 and 24031 have the most average Master or Professional percent
```
